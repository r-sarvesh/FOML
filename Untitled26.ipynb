```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, IsolationForest
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    classification_report, confusion_matrix,
    precision_recall_curve, average_precision_score,
    roc_auc_score, f1_score
)
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
import xgboost as xgb
import warnings
warnings.filterwarnings('ignore')

RANDOM_STATE = 42
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
df = pd.read_csv('creditcard.csv')

print("Shape:", df.shape)
print("\nClass distribution:")
print(df['Class'].value_counts())
print(f"\nFraud %: {df['Class'].mean() * 100:.4f}%")

# Visual: class imbalance
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Class distribution bar
counts = df['Class'].value_counts()
axes[0].bar(['Legitimate', 'Fraud'], counts.values,
            color=['#1D9E75', '#D85A30'], edgecolor='white')
axes[0].set_title('Class Distribution (raw)')
axes[0].set_ylabel('Count')
for i, v in enumerate(counts.values):
    axes[0].text(i, v + 500, f'{v:,}', ha='center', fontweight='bold')

# Amount distribution by class
df[df['Class'] == 0]['Amount'].clip(upper=500).hist(
    ax=axes[1], bins=50, color='#1D9E75', alpha=0.7, label='Legitimate')
df[df['Class'] == 1]['Amount'].clip(upper=500).hist(
    ax=axes[1], bins=50, color='#D85A30', alpha=0.7, label='Fraud')
axes[1].set_title('Transaction Amount (clipped at $500)')
axes[1].set_xlabel('Amount ($)')
axes[1].legend()

# Correlation heatmap of V features
corr = df[['V1','V2','V3','V4','V5','Amount','Class']].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='RdYlGn',
            center=0, ax=axes[2], cbar_kws={'shrink': 0.8})
axes[2].set_title('Feature Correlations')

plt.tight_layout()
plt.savefig('eda_plots.png', dpi=120, bbox_inches='tight')
plt.show()

print("\nFraud transaction stats:")
print(df[df['Class'] == 1]['Amount'].describe())
```

```python
# Scale 'Amount' and 'Time' (V1–V28 are already PCA'd and scaled)
scaler = StandardScaler()
df['Amount_scaled'] = scaler.fit_transform(df[['Amount']])
df['Time_scaled'] = scaler.fit_transform(df[['Time']])

# Drop original columns
df.drop(['Amount', 'Time'], axis=1, inplace=True)

# Features and target
X = df.drop('Class', axis=1)
y = df['Class']

# Stratified split to preserve fraud ratio
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
)

print(f"Train size: {X_train.shape[0]:,} | Test size: {X_test.shape[0]:,}")
print(f"Train fraud count: {y_train.sum()} | Test fraud count: {y_test.sum()}")
```

```python
smote = SMOTE(random_state=RANDOM_STATE, k_neighbors=5)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

print("Before SMOTE:", dict(pd.Series(y_train).value_counts()))
print("After SMOTE: ", dict(pd.Series(y_train_res).value_counts()))

# Visualize SMOTE effect
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
for ax, (label, counts, title) in zip(axes, [
    (y_train.value_counts(), y_train.value_counts(), 'Before SMOTE'),
    (y_train_res, pd.Series(y_train_res).value_counts(), 'After SMOTE'),
]):
    pd.Series(y_train_res if title == 'After SMOTE' else y_train)\
      .value_counts().plot.bar(ax=ax, color=['#1D9E75', '#D85A30'],
                               edgecolor='white', rot=0)
    ax.set_xticklabels(['Legitimate', 'Fraud'])
    ax.set_title(title)
    ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
def evaluate_model(name, y_true, y_pred, y_proba):
    """Print full evaluation with PR curve data."""
    print(f"\n{'='*50}")
    print(f"  {name}")
    print(f"{'='*50}")
    print(classification_report(y_true, y_pred, target_names=['Legitimate', 'Fraud']))

    auprc = average_precision_score(y_true, y_proba)
    roc   = roc_auc_score(y_true, y_proba)
    f1    = f1_score(y_true, y_pred)

    print(f"  AUPRC (Avg Precision): {auprc:.4f}")
    print(f"  ROC-AUC:               {roc:.4f}")
    print(f"  F1 Score (fraud):      {f1:.4f}")

    return auprc, roc, f1

# Train
lr = LogisticRegression(class_weight='balanced', max_iter=1000,
                         random_state=RANDOM_STATE)
lr.fit(X_train_res, y_train_res)
lr_proba = lr.predict_proba(X_test)[:, 1]
lr_pred  = lr.predict(X_test)

lr_auprc, lr_roc, lr_f1 = evaluate_model(
    'Logistic Regression (Baseline)', y_test, lr_pred, lr_proba)
```

```python
rf = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    class_weight='balanced',
    random_state=RANDOM_STATE,
    n_jobs=-1
)
rf.fit(X_train_res, y_train_res)
rf_proba = rf.predict_proba(X_test)[:, 1]
rf_pred  = rf.predict(X_test)

rf_auprc, rf_roc, rf_f1 = evaluate_model(
    'Random Forest', y_test, rf_pred, rf_proba)
```

```python
# Calculate scale_pos_weight for class imbalance
neg = (y_train == 0).sum()
pos = (y_train == 1).sum()
scale = neg / pos

xgb_model = xgb.XGBClassifier(
    n_estimators=200,
    max_depth=6,
    learning_rate=0.05,
    scale_pos_weight=scale,     # handles imbalance internally
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric='aucpr',
    random_state=RANDOM_STATE,
    use_label_encoder=False
)
xgb_model.fit(
    X_train, y_train,           # use original (not SMOTE) for XGBoost
    eval_set=[(X_test, y_test)],
    verbose=False
)
xgb_proba = xgb_model.predict_proba(X_test)[:, 1]
xgb_pred  = xgb_model.predict(X_test)

xgb_auprc, xgb_roc, xgb_f1 = evaluate_model(
    'XGBoost (scale_pos_weight)', y_test, xgb_pred, xgb_proba)
```

```python
iso = IsolationForest(
    n_estimators=200,
    contamination=0.002,    # approx fraud rate
    random_state=RANDOM_STATE,
    n_jobs=-1
)
iso.fit(X_train)

# IsolationForest: -1 = anomaly, 1 = normal → map to 0/1
iso_pred_raw = iso.predict(X_test)
iso_pred = np.where(iso_pred_raw == -1, 1, 0)

# Score: lower = more anomalous → negate for proba-like score
iso_scores = -iso.score_samples(X_test)
# Normalize to [0,1]
iso_proba = (iso_scores - iso_scores.min()) / (iso_scores.max() - iso_scores.min())

iso_auprc, iso_roc, iso_f1 = evaluate_model(
    'Isolation Forest (Unsupervised)', y_test, iso_pred, iso_proba)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# --- Precision-Recall Curves ---
ax = axes[0]
models = {
    'Logistic Regression': (lr_proba,  '#888780'),
    'Random Forest':        (rf_proba,  '#1D9E75'),
    'XGBoost':              (xgb_proba, '#D85A30'),
    'Isolation Forest':     (iso_proba, '#7F77DD'),
}
for name, (proba, color) in models.items():
    prec, rec, _ = precision_recall_curve(y_test, proba)
    ap = average_precision_score(y_test, proba)
    ax.plot(rec, prec, label=f'{name} (AUPRC={ap:.3f})',
            color=color, linewidth=2)

# Baseline (random classifier)
ax.axhline(y=y_test.mean(), color='gray', linestyle='--',
           alpha=0.5, label=f'Random (AUPRC={y_test.mean():.3f})')
ax.set_xlabel('Recall', fontsize=12)
ax.set_ylabel('Precision', fontsize=12)
ax.set_title('Precision-Recall Curves', fontsize=13, fontweight='bold')
ax.legend(fontsize=9)
ax.set_xlim([0, 1])
ax.set_ylim([0, 1.02])

# --- Confusion Matrices (XGBoost) ---
ax2 = axes[1]
cm = confusion_matrix(y_test, xgb_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='YlOrRd',
            xticklabels=['Predicted Legit', 'Predicted Fraud'],
            yticklabels=['Actual Legit', 'Actual Fraud'],
            ax=ax2, cbar=False)
ax2.set_title('XGBoost Confusion Matrix', fontsize=13, fontweight='bold')
tn, fp, fn, tp = cm.ravel()
ax2.set_xlabel(
    f'TP={tp} | FP={fp} | FN={fn} | TN={tn:,}', fontsize=10)

plt.tight_layout()
plt.savefig('pr_curves_and_cm.png', dpi=120, bbox_inches='tight')
plt.show()
```

```python
feat_imp = pd.Series(xgb_model.feature_importances_,
                     index=X.columns).sort_values(ascending=False)

plt.figure(figsize=(10, 5))
feat_imp.head(15).plot.bar(color='#D85A30', edgecolor='white')
plt.title('Top 15 Features — XGBoost', fontsize=13, fontweight='bold')
plt.ylabel('Importance Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('feature_importance.png', dpi=120, bbox_inches='tight')
plt.show()
```

```python
import joblib

# Summary table
summary = pd.DataFrame({
    'Model':  ['Logistic Regression', 'Random Forest', 'XGBoost', 'Isolation Forest'],
    'AUPRC':  [lr_auprc, rf_auprc, xgb_auprc, iso_auprc],
    'ROC-AUC':[lr_roc,   rf_roc,   xgb_roc,   iso_roc],
    'F1':     [lr_f1,    rf_f1,    xgb_f1,    iso_f1],
}).sort_values('AUPRC', ascending=False).reset_index(drop=True)

print("\n===== Model Comparison =====")
print(summary.to_string(index=False))

# Save best model
joblib.dump(xgb_model, 'fraud_detector_xgboost.pkl')
joblib.dump(scaler,    'scaler.pkl')
print("\nModels saved: fraud_detector_xgboost.pkl & scaler.pkl")
```
